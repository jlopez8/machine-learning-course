This is done with a 'remove and replace' procedure.

```python
# This re-orders the current columns list of the dataframe to a new order as a list.
cols = list(titanic_data.columns)

 # remove from list operation
cols.remove('survived')

# list operation insert element at index
cols.insert(0, 'survived') 
cols

# this next command will re-index the columns by callling them out in specified order
titanic_data = titanic_data.reindex(columns=cols)
```

## Response Variable visualizations

```python
import seaborn as sns 
```

We can visualize groups of variables in relation to the response variable.

```python
titanic_data.columns
sns.catplot(data=titanic_data, kind='bar', x='sex', y='parch', hue='survived')
```

Interpretation  

For females, when the par,child proportion is high (they tend to have more parents and children on board), they tend to surive less - this is the blue category. Fewer parents and children on board equates to more survival.  

In males, the parent and child numbers being lower in proportion says they're less likely to have survived the Titanic. 

## Violin Plot

```python
sns.violinplot(data=titanic_data, x='sex', y='age', hue='survived', split=True)
```

Violin plots show essentially a shape of histogram data as a shape. For Females, out of those that survived, older females tended to persist. While for those who died, mostly younger females constitute that population including an overwelming amount of girls.  

For males, predominantly younger males died but not children (boys). Mostly thoser in their early 20s. For the males that survived, most tended to be either in their 20s as well or young boys (~5 y/o).

## Correlation

```python
corr = titanic_data.corr(numeric_only=True)

## since we don't know... we can define the mast as the upper triangle of 1's with
## same shape and size as the correlation array.
## mask is applied when we have missing values. 
mask = np.triu(np.ones_like(corr, dtype=bool))


plt.subplots(figsize=(10,8))
# center of 0 is very important. Corr is defined as +1, -1 values with center 0. The 
# heatmap should be centered at the same vlue.
sns.heatmap(
    corr, mask=mask, cmap='seismic', 
    center=0, square=True, linewidth=0.5,
    cbar_kws={'shrink':0.5} #optional visualization modifications
) 
```

Post analysis show that **fare** and **survived** have a high positive correlation. This means the richer people / people with a higher fared ticket were more likely to survive. 

We see **fare** and **pclass** are negatively correlated. This means we have a higher fair corresponding to a lower class number (which really means the premier or higher valued calss 1 class > 2 class > 3 class and so on).
